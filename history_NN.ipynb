# Perceptrons

In Prof. Ng's week 3 class he introduced the concept of a simple neural network using different binary bitwise operations. What he didn't mention was why ** XOR ** was a good example for this introduction. In 1969, Minsky and Papert published a book [Perceptrons](https://mitpress.mit.edu/books/perceptrons) that proved that you can't find a line to separate two classes of XOR bitwise operations, and that almost killed the research in Artificial Neural Networks. The reason was that Perceptrons was very simple and only had one single layer which can only solve bitwise operation problems such as ** OR ** and ** AND **. 

Can you find a line that separates yellow dots from purple dots?



```python
import numpy as np
import matplotlib.pyplot as plt 

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])

# plot the (testing) classification data
plt.style.use("ggplot")
plt.figure()
plt.title("XOR - Non-linear classification")
plt.xlabel('x1')
plt.ylabel('x2')

plt.scatter(X[:, 0], X[:, 1], marker="o", c=np.squeeze(y), s=200)

print('You cannot find a line to separate these two classes, which is the output of XOR bitwise operation')
```
